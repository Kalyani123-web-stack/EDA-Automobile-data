```python
#delete columns

df.drop(['symboling','normalized-losses','aspiration','num-of-doors','engine-location','fuel-system','bore','stroke'],axis=1,inplace=True)
```

```python
#rename columns

df=df.rename(columns={'make':'car-brand'})
print(df)
```

```python
#check columns list after renameing column.
print(df.columns)
```

```python
# count particular column values
df['car-brand'].value_counts()
```

```python
#capitalize column
def uppercase(df):
  df['car-brand']= df['car-brand'].apply(lambda x: x.upper())
  return df

df=uppercase(df)
print(df)
```

```python
#Descriptive statistics for numeric columns.
df.describe()
```

```python
# Check for non-numeric values in the 'price' column

print(df['price'].nunique())
```

```python
# Convert 'price' to numeric, setting errors='coerce' to handle non-numeric values
df['price']=pd.to_numeric(df['price'],errors='coerce')
```

```python
#fill NaN values with a default value
df['price']=df['price'].fillna(0)
```

```python
#calculate average price by car-brand

avg_price_brand=df.groupby('car-brand')['price'].mean()
print(avg_price_brand)
```

```python
avg_price_brand=df.groupby('car-brand')['price'].mean().sort_values(ascending=True)

import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.xlabel('Car-brand',color='blue')
plt.xticks(rotation=90)
plt.ylabel('Price',color='blue')
plt.title('Bar plot')
plt.bar(avg_price_brand.index,avg_price_brand.values,color='red')
plt.show()
```

Observation1:The bar plot shows that,
1.The **JAGUAR** & **MERCEDES-BENZ** are the top car brands which have highest price around 35000.   
2. The **ISUZU** car has lowest price upto 5000.
3. Certain cars like **SAAB,AUDI,PEUGOT,VOLVO** also in demanding,have moderate prices nearby 20000.



```python
#calculate average-price by using fuel-type
avg_price_fuel_type=df.groupby('fuel-type')['price'].mean()
print(avg_price_fuel_type)
```

```python
plt.xlabel('Fuel-type',color='blue')
plt.xticks(rotation=45)
plt.ylabel('Price',color='blue')
plt.title('Scatter plot')
plt.scatter(avg_price_fuel_type.index,avg_price_fuel_type.values,color='green',s=100)
plt.grid(True)
plt.show()
```

Observation 2:The diesel car are costly (above 15500) as compared to gas.

```python
# Convert 'horsepower' and 'curb-weight' to numeric
df['horsepower']=pd.to_numeric(df['horsepower'],errors='coerce')
df['curb-weight']=pd.to_numeric(df['curb-weight'],errors='coerce')
```

```python
df['power_to_weight_ratio']=df['horsepower']/df['curb-weight']
print(df['power_to_weight_ratio'])
```

```python
top_10_cars=df.sort_values(by='power_to_weight_ratio',ascending=False).head(10)
top_10_cars=top_10_cars[['car-brand','power_to_weight_ratio']]
print(top_10_cars)
```

```python
filtered_top_10 = top_10_cars.loc[top_10_cars['power_to_weight_ratio'] > 0.080]

for index, row in filtered_top_10.iterrows():
    car_brand = row['car-brand']
    print(f'{car_brand} better perform in term of speed & acceleration')
```

```python
#RPM per Horsepower:It could be an indicator of engine efficiency.
```

```python
df['peak-rpm']=pd.to_numeric(df['peak-rpm'],errors='coerce')
df['horsepower']=pd.to_numeric(df['horsepower'],errors='coerce')
```

```python
df['rpm_per_horsepower']=df['peak-rpm']/df['horsepower']
print(df['rpm_per_horsepower'])
```

```python
df[['car-brand','rpm_per_horsepower']].head(10)
```

```python
top_10_cars=df.sort_values(by='rpm_per_horsepower',ascending=True)
top_10_cars=top_10_cars[['car-brand','rpm_per_horsepower']].head(10)
print(top_10_cars)
```

```python
#JAGUAR vehicles in have a very low RPM relative to horsepower, indicating they are designed for efficient power output at lower RPMs.
# MERCEDES-BENZ: The values of 24.456522 is a balance between RPM and horsepower, indicating a mix of efficiency and performance.
# NISSAN: With a value of 26.000000, NISSAN vehicles is  slightly higher in terms of RPM per horsepower compared to the other brands listed.
```

```python
# A lower rpm_per_horsepower typically indicates higher efficiency.
avg_rpm_per_horsepower = df.groupby('car-brand')['rpm_per_horsepower'].mean()
print(avg_rpm_per_horsepower)
```

```python
# Lowest RPM per Horsepower:

# JAGUAR: 24.353747
# MERCURY: 28.571429
# PORSCHE: 28.786813
# These brands have low rpm_per_horsepower, indicating that their engines generate power more efficiently at lower RPMs.

# Highest RPM per Horsepower:

# CHEVROLET: 86.845238
# HONDA: 73.181178
# ISUZU: 67.844933
# These brands have higher rpm_per_horsepower, meaning they require higher RPMs to generate the same amount of horsepower
```

```python
# Efficiency Index:combine compression-ratio and horsepower to create an "Efficiency Index," where you assess how much power the engine can produce relative to the compression ratio.
```

```python
df['efficiency_index']=df['compression-ratio']/df['horsepower']
print(df['efficiency_index'])
```

```python
df[['car-brand','horsepower','efficiency_index']].head()
```

```python
avg_efficiency_index = df.groupby('car-brand')['efficiency_index'].mean()
print(avg_efficiency_index)
```
